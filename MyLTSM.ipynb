```python
def activation(function,x):
    e = 2.718281828459045 #자연상수:2.71828182845904523536
    if(function == 'Sigmoid'):
        return (e**x)/((e**x)+1) #C언어에서는 **이 없어 **도 구현
    if(function == 'Tanh'):
        return ((e**x) - (e**(-x))) / ((e**x) + (e**(-x)))
    if(function == 'ReLU'):
        if (x<=0):
            return 0
        return x
```

```python
def matrix_dot(A, B):
    if type(A[0]) == list:
        if len(A[0]) != len(B):
            raise ValueError("행렬 곱 연산이 불가능합니다.")
        
        result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
        #print(np.array(result).shape)
        for i in range(len(A)):
            for j in range(len(B[0])):
                dot_product = 0
                for k in range(len(A[0])):
                    dot_product += A[i][k] * B[k][j]
                result[i][j] = dot_product
        return result

    if type(A[0]) == int:
        if len(A) != len(B):
            raise ValueError("행렬 곱 연산이 불가능합니다.")
        result = []
        for i in range(len(B)):
            for j in range(len(B[0])):
                result.append(A[0]*B[0][j])
        result = [result]
        #print(np.array(result).shape) 
        return result
```

```python
def element_sum(X,w,Y,v,b):
    dot_X = matrix_dot(X,w)
    dot_Y = matrix_dot(Y,v)
    result = [0 for i in range(len(b[0]))]
    for i in range(len(b[0])):
        result[i] = dot_X[0][i] +dot_Y[0][i] + b[0][i]
    result = [result]
    return result 
```
